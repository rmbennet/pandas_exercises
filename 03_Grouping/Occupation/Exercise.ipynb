### Step 3. Assign it to a variable called users.

```python
users = pd.read_table(filepath_or_buffer=filename,sep='|')
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')[['age']].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

Redefine the gender variable:
$$
\begin{equation*}
\text{Gender}=
\begin{cases}
1&\text{if Male}\\
0&\text{Otherwise}\\
\end{cases}
\end{equation*}
$$

```python
transform_gender = lambda x: 1 if x=='M' or x==1 else 0
users['gender']=users['gender'].apply(transform_gender)
subset_5 = users.groupby('occupation').sum()[['gender']].rename(columns={'gender':'total_male'})
occupation_totals = users['occupation'].value_counts()

subset_5 = subset_5.assign(prop_male = subset_5['total_male']/occupation_totals)
```
